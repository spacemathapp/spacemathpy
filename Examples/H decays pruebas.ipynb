```python
import spacemathpy as spm
```

```python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib qt
%matplotlib inline
import sympy as sp
sp.init_printing()
```

```python
s = 'symbol'
v = 'value'
ghtt,ghbb = sp.symbols('g_{htt},g_{hbb}',positive=True)
```

```python
xt = lambda mS: 4*spm.mt[s]**2/(mS**2)
xb = lambda mS: 4*spm.mb[s]**2/(mS**2)
At = lambda mS: sp.asin(sp.sqrt((mS**2)/(4*spm.mt[s]**2)))**2
Ab = lambda mS: -(sp.S(1)/4)*(sp.log((1+sp.sqrt(1-xb(mS)))/(1-sp.sqrt(1-xb(mS))))-sp.I*sp.pi)**2
```

```python

Fb = lambda mS:-2*xb(mS)*(1+(1-xb(mS))*Ab(mS))
Ft = lambda mS:-2*xt(mS)*(1+(1-xt(mS))*At(mS))
```

```python
AHgg = lambda ghtt,ghbb,mS: 2*spm.mW[s]*((ghtt/(spm.mt[s])*Ft(mS))+(ghbb/(spm.mb[s])*Fb(mS)))
```

```python
Whgg = ((spm.αs[s]**2*spm.mh[s]**3)/(512*spm.mW[s]**2*sp.pi**3 ))*(
    AHgg(ghtt,ghbb,spm.mh[s]))**2
Whgg
```

```python
res = Whgg.expand().collect([ghtt,ghbb],sp.factor)
res
```

```python
whggs = sp.symbols('\Gamma_{hgg}',positive=True)
sols = sp.solve(sp.Eq(res,whggs),ghtt)
```
